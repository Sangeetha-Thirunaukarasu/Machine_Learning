```python
#splitting dependent and independent variable
X = data.iloc[:,2:-1]
y = data.loc[:,['diagnosis']]
#X.drop('Unnamed: 32',axis=1,inplace=True) #dropping the column Unnamed-32 which hs no non null values
X
#y


```

## splitting

```python
#splitting the data set for train and test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify = y)
y_train.value_counts()
#y_test
```

## balancing - using Oversampling

```python
#we have imabalanced data set. Oversampling to make it balanced since we have less no.of records

#importing smote
from imblearn.over_sampling import SMOTE
#creating object for class smote
smt = SMOTE()
X_train_sam, y_train_sam = smt.fit_resample(X_train,y_train)
y_train_sam.value_counts()
```

```python
#before scaling
X_train.boxplot()
#plt.plot(X_train)
plt.show()
```

## Scaling

```python
#features scaling - standardScaler
from sklearn.preprocessing import StandardScaler
mms= StandardScaler()
mms.fit(X_train)

X_train_scaled = mms.transform(X_train_sam)
X_test_scaled = mms.transform(X_test)
```

```python

#plotting after scaling
plt.boxplot(X_train_scaled,vert=True)
plt.show()
```

## Feature Extraction
## PCA - to reduce the dimensionality of the data
#reduce highly correlated variables


```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
pca.fit(X_train_scaled)
```

```python
pca.n_components_ #here number of components having 95% of variance are 10
```

```python
##Transform both train and test data
X_tra = pca.transform(X_train_scaled)
X_tes = pca.transform(X_test_scaled)
```

```python
print('total training points:----')
print(X_tra.shape)

print('datapoints for target variable' )
print(y_train_sam.value_counts())

print('Test data:-----')
print(X_tes.shape)

print(y_test.value_counts())
```

## Feature Selection

```python
#ANOVA feature selection technique
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif,chi2
fs = SelectKBest(score_func = f_classif, k=15)
#fs = SelectKBest(score_func = chi2, k=15)
fs.fit(X_train_scaled,y_train_sam)
X_train_select = fs.transform(X_train_scaled)
X_test_select = fs.transform(X_test_scaled)
X_train_select.shape
```

## Ensemble learning

```python
from sklearn.ensemble import RandomForestClassifier
scores =[]
for i in range(5,11):
    print('n_estimator: ',i)
    for j in range(5,11):
        rfc = RandomForestClassifier(n_estimators =i, criterion='entropy', max_depth=j)
        rfc.fit(X_train_select,y_train_sam)
        rf_pred = rfc.predict(X_test_select)
        print("------------- ---------- ")
        print('maximum depth :',j) 
        print('auc_roc_score :',roc_auc_score(y_test,rf_pred))
        print('classification_report : ')
        print(classification_report(y_test,rf_pred))
    
    scores.append(accuracy_score(y_test,rf_pred))
    #score.append(sco)
print(" ")
#print('accuracy: ',scores)
#pred    
```

# Hyperparameter tuning

```python
#C -regularization parameter - strength of regularisation is inversely proportional to 'C'
param_distributions={'C':[0.1,1,10,100,1000],'kernel':['rbf','linear','sigmoid','poly']}

from sklearn.model_selection import GridSearchCV,RandomizedSearchCV,cross_val_score

svc_hyper = SVC()
rcv = RandomizedSearchCV(svc_hyper,param_distributions,cv=5,n_iter=6,random_state=0)
rcv.fit(X_train_select,y_train_sam)
```

```python
param_grid={'C':[0.1,1,10,100,1000],'kernel':['rbf','linear','sigmoid','poly']}
gridcv = GridSearchCV(svc,param_grid,cv=5)
gridcv.fit(X_train_select,y_train_sam)
```
